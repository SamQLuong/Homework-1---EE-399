```python
#https://github.com/SamQLuong/Homework-1---EE-399

# imports the libraries needed for this project
import numpy as np
import scipy.optimize as opt
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 31) # X axis data
y = np.array([30, 35, 33, 32, 34, 37, 39, 38, 36, 36, 37, 39, 42, 45, 45, 41, 
              40, 39, 42, 44, 47, 49, 50, 49, 46, 48, 50, 53, 55, 54, 53]) # Y axis Data

# Function that returns the least square error based on the parameters
# p is the array of guessed parameters, x is the x data, and y is the y data
def bestfit(p, x, y):
    
    e = np.sqrt(np.sum((p[0] * np.cos(p[1]*x) + p[2] * x + p[3] - y)**2) / 31)
    
    return e

# the guessing data points for the four missing parameters
p = np.array([3, (1 * np.pi) / 5, 15, 31])

# minimizes and finds the best fitting model and returns the parameters
result = opt.minimize(bestfit, p, args = (x, y), method = 'Nelder-Mead')

# seperates the parameters from the other results
param = result.x

# print individual parameters
print('A = ', param[0])
print('B = ', param[1])
print('C = ', param[2])
print('D = ', param[3])

xfit = np.arange(0, 31.01, 0.01) # creates the x axis for plotting
yfit = param[0] * np.cos(param[1] * xfit) + param[2] * xfit + param[3] # creates the y axis based on the new parameters and x data

#plots the original data points and the new model
plt.title('Curve Fitting')
plt.plot(x, y, 'ko')
plt.plot(xfit, yfit, 'r-')
plt.xlabel('X Data')
plt.ylabel('Y Data')
plt.show()
```

```python
# The function that returns the least square error based on the 4 parameters, A, B, C, D
# the x and y values are given from the last part
def loss_fun(A, B, C, D, x, y):
    e = np.sqrt(np.sum((A * np.cos(B*x) + C * x + D - y)**2) / 31)
    return e

# fixed values of A, B, C, and D from the last part's model
fixed_A = 2.1717
fixed_B = 0.909
fixed_C = 0.732
fixed_D = 31.45

# the sweeping of parameters C and D
# range depends on the fixed values and number of values set to 10
C1 = np.linspace(0.5, 0.9, 10)
D1 = np.linspace(20, 40, 10)

# creates the 2D Loss Landscape Array by using two for loops and adding the
# least sqaure error value
count1 = 0
loss_val1 = np.zeros((len(C1), len(D1)))
for i in range(len(C1)):
    for j in range(len(D1)):
        loss_val1[i][j] = loss_fun(fixed_A, fixed_B, C1[i], D1[j], x, y)
        if loss_val1[i][j] < 2:
            count1 += 1

# the sweeping of parameters A and D
# range depends on the fixed values and number of values set to 10
A2 = np.linspace(0, 4, 10)
D2 = np.linspace(20, 40, 10)

# creates the 2D Loss Landscape Array by using two for loops and adding the
# least sqaure error value
# a counter is used to find the number of least square erros less than 2
count2 = 0
loss_val2 = np.zeros((len(A2), len(D2)))
for i in range(len(A2)):
    for j in range(len(D2)):
        loss_val2[i][j] = loss_fun(A2[i], fixed_B, fixed_C, D2[j], x, y)
        if loss_val2[i][j] < 2:
            count2 += 1

# the sweeping of parameters A and B
# range depends on the fixed values and number of values set to 10
A3 = np.linspace(0, 4, 10)
B3 = np.linspace(0.7, 1.1, 10)

# creates the 2D Loss Landscape Array by using two for loops and adding the
# least sqaure error value
# a counter is used to find the number of least square erros less than 2
count3 = 0
loss_val3 = np.zeros((len(A3), len(B3)))
for i in range(len(A3)):
    for j in range(len(B3)):
        loss_val3[i][j] = loss_fun(A3[i], B3[j], fixed_C, fixed_D, x, y)
        if loss_val3[i][j] < 2:
            count3 += 1

# the sweeping of parameters B and C
# range depends on the fixed values and number of values set to 10
B4 = np.linspace(0.7, 1.1, 10)
C4 = np.linspace(0.5, 0.9, 10)

# creates the 2D Loss Landscape Array by using two for loops and adding the
# least sqaure error value
# a counter is used to find the number of least square erros less than 2
count4 = 0
loss_val4 = np.zeros((len(B4), len(C4)))
for i in range(len(B4)):
    for j in range(len(C4)):
        loss_val4[i][j] = loss_fun(fixed_A, B4[i], C4[j], fixed_D, x, y)
        if loss_val4[i][j] < 2:
            count4 += 1

# the sweeping of parameters B and D
# range depends on the fixed values and number of values set to 10
B5 = np.linspace(0.7, 1.1, 10)
D5 = np.linspace(20, 40, 10)

# creates the 2D Loss Landscape Array by using two for loops and adding the
# least sqaure error value
# a counter is used to find the number of least square erros less than 2
count5 = 0
loss_val5 = np.zeros((len(B5), len(D5)))
for i in range(len(B5)):
    for j in range(len(D5)):
        loss_val5[i][j] = loss_fun(fixed_A, B5[i], fixed_C, D5[j], x, y)
        if loss_val5[i][j] < 2:
            count5 += 1

# the sweeping of parameters A and C
# range depends on the fixed values and number of values set to 10
A6 = np.linspace(1, 3, 10)
C6 = np.linspace(0.5, 0.9, 10)

# creates the 2D Loss Landscape Array by using two for loops and adding the
# least sqaure error value
# a counter is used to find the number of least square erros less than 2
count6 = 0
loss_val6 = np.zeros((len(A6), len(C6)))
for i in range(len(A6)):
    for j in range(len(C6)):
        loss_val6[i][j] = loss_fun(A6[i], fixed_B, C6[j], fixed_D, x, y)
        if loss_val6[i][j] < 2:
            count6 += 1
            
print("number of least square errors less than 2 for the 6 combinations", count1,count2,count3,count4,count5,count6)

# uses the 2D loss arrays and plugs it into the pcolor code
# creates a graph that shows darker color spots as near to 0 least squared error
# includes axis labels, titles, and colorbar
fig, axs = plt.subplots(2, 3, figsize=(10, 8))
fig.suptitle('2D Loss Landscape with 2 Sweeping Parameters')

clb1 = axs[0,0].pcolor(loss_val1, edgecolors = 'k', linewidths = 2)
axs[0,0].set_xlabel("Sweep Parameter C")
axs[0,0].set_ylabel("Sweep Parameter D")
fig.colorbar(clb1, ax=axs[0,0])

clb2 = axs[0,1].pcolor(loss_val2, edgecolors = 'k', linewidths = 2)
axs[0,1].set_xlabel("Sweep Parameter A")
axs[0,1].set_ylabel("Sweep Parameter D")
fig.colorbar(clb2, ax=axs[0,1])

clb3 = axs[0,2].pcolor(loss_val3, edgecolors = 'k', linewidths = 2)
axs[0,2].set_xlabel("Sweep Parameter A")
axs[0,2].set_ylabel("Sweep Parameter B")
fig.colorbar(clb3, ax=axs[0,2])

clb4 = axs[1,0].pcolor(loss_val4, edgecolors = 'k', linewidths = 2)
axs[1,0].set_xlabel("Sweep Parameter B")
axs[1,0].set_ylabel("Sweep Parameter C")
fig.colorbar(clb4, ax = axs[1,0])

clb5 = axs[1,1].pcolor(loss_val5, edgecolors = 'k', linewidths = 2)
axs[1,1].set_xlabel("Sweep Parameter B")
axs[1,1].set_ylabel("Sweep Parameter D")
fig.colorbar(clb5, ax = axs[1,1])

clb6 = axs[1,2].pcolor(loss_val6, edgecolors = 'k', linewidths = 2)
axs[1,2].set_xlabel("Sweep Parameter A")
axs[1,2].set_ylabel("Sweep Parameter C")
fig.colorbar(clb6, ax = axs[1,2])

plt.tight_layout()
plt.show()
```

```python
new_X = x[:20] # first 20 x data points as training data
new_Y = y[:20] # first 20 y data points as training data

line_p = np.polyfit(new_X, new_Y, 1) # returns the linear model with the training data

para_p = np.polyfit(new_X, new_Y, 2) # returns the parabola model with the training data

nineteen_p = np.polyfit(new_X, new_Y, 19) # returns the 19th degree model with the training data

# returns the least squared error of the linear model with the parameters of
# p: array of parameters, x: x data, y: y data 
def errorline(p, x, y):
    e = np.sqrt(np.sum((p[0] * x + p[1] - y)**2) / len(y))
    return e

# returns the least squared error of the parabola model with the parameters of
# p: array of parameters, x: x data, y: y data 
def errorpara(p, x, y):
    e = np.sqrt(np.sum((p[0] * x ** 2 + p[1] * x ** 2 + p[2] - y)**2) / len(y))
    return e

# returns the least squared error of the 19th degree model with the parameters of
# p: array of parameters, x: x data, y: y data 
def errornineteen(p, x, y):
    e = np.sqrt(np.sum((np.polyval(p, x) - y)**2) / len(y))
    return e

# print out the least squared error for the training data and with the testing data
# in this case the testing data is the last 10 data points.
print("least squared error for a line (first 20) = ", errorline(line_p, new_X, new_Y))
print("least squared error for a parabola (first 20) = ", errorpara(para_p, new_X, new_Y))
print("least squared error for a nineteenth polynomial (first 20) = ", errornineteen(nineteen_p, new_X, new_Y))
print("")
print("least squared error for a line (last 10) = ", errorline(line_p, x[-10:], y[-10:]))
print("least squared error for a parabola (last 10) = ", errorpara(para_p, x[-10:], y[-10:]))
print("least squared error for a nineteenth polynomial (last 10) = ", errornineteen(nineteen_p, x[-10:], y[-10:]))

# creates 3 plots with each model where the black dots are the training data
# the blue dots are the testing data
# and the red line is the model from the training data
fig, axs = plt.subplots(1,3, figsize = (10,8))

fig.suptitle("Training Different Models with the First 20 Data Points")

xfit =  np.arange(0, 31.01, 0.01)

yline1 = line_p[0] * xfit + line_p[1] # linear model equation

ypara1 = para_p[0] * xfit ** 2 + para_p[1] * xfit + para_p[2] # parabola model equation

ynine1 = np.polyval(nineteen_p, new_X) # 19th degree model equation

axs[0].plot(new_X, new_Y, 'ko', label = 'Training Data')
axs[0].plot(x[-10:], y[-10:], 'bo', label = 'Testing Data')
axs[0].plot(xfit, yline1, 'r-', label = 'Linear Model')
axs[0].set_xlabel("X Data")
axs[0].set_ylabel("Y Data")
axs[0].legend()

axs[1].plot(new_X, new_Y, 'ko', label = 'Training Data')
axs[1].plot(x[-10:], y[-10:], 'bo', label = 'Testing Data')
axs[1].plot(xfit, ypara1, 'r-', label = 'Parabola Model')
axs[1].set_xlabel("X Data")
axs[1].set_ylabel("Y Data")
axs[1].legend()

axs[2].plot(new_X, new_Y, 'ko', label = 'Training Data')
axs[2].plot(x[-10:], y[-10:], 'bo', label = 'Testing Data')
axs[2].plot(new_X, ynine1, 'r-', label = '19th Poly Model')
axs[2].set_xlabel("X Data")
axs[2].set_ylabel("Y Data")
axs[2].legend()

plt.tight_layout()
plt.show()
```
